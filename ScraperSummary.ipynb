```python
import re 
import json
import csv
from io import StringIO
from bs4 import BeautifulSoup
import requests
```

```python
f = open("ListaSimboli.txt", "r")
for s in f: 
    
    stock= s.strip()
    
    url_stats = 'https://finance.yahoo.com/quote/{}?p={}'

    response = requests.get(url_stats.format(stock, stock))
    soup = BeautifulSoup(response.text, 'html.parser')
    print(soup.title.text)

    close = soup.find('span', {'class':'Trsdu(0.3s) Fw(b) Fz(36px) Mb(-4px) D(ib)'}).text

    print(close)

    dividend = soup.find_all('td', {'class': 'Ta(end) Fw(600) Lh(14px)'})  

    data = {}
    data ["Symbol"] = stock 
    data ["Close"] = close 
    data ["Open"] = dividend[1].get_text() 
    data ["Volume"] = dividend[6].get_text() 
    data ["Avg_Volume"] = dividend[7].get_text() 
    data ["Market_cap"] = dividend[8].get_text() 
    data ["PE_Ratio"] = dividend[10].get_text() 
    data ["EPS"] = dividend[11].get_text() 
    data ["Dividend"] = dividend[13].get_text() 
    data ["Target_prize"] = dividend[15].get_text() 

    print(data)
    
    with open("Summary"+stock+".json", "w") as jsonFile:
        jsonFile.write(json.dumps(data, indent=3))

    jsonFile.close()
f.close()
```
