```python
import re 
import json
import csv
from io import StringIO
from bs4 import BeautifulSoup
import requests
```

```python
f = open("ListaSimboli.txt", "r")
jsonFile = open("Summary.json", "w")
jsonFile.write("[")
for s in f: 
    
    stock= s.strip()
    
    url_stats = 'https://finance.yahoo.com/quote/{}?p={}'

    response = requests.get(url_stats.format(stock, stock))
    soup = BeautifulSoup(response.text, 'html.parser')
    
    #close = soup.find('span', {'class':'Trsdu(0.3s) Fw(b) Fz(36px) Mb(-4px) D(ib)'}).text
    try:
        close = soup.find('div', {'class':'D(ib) Mend(20px)'}).find_all('span')[0].text
    except:
        print("Azienda non trovata" + stock)
        continue
    
           
    print(close)

    dividend = soup.find_all('td', {'class': 'Ta(end) Fw(600) Lh(14px)'})  

    data = {}
    data ["Symbol"] = stock
    
    if close != 'N/A' :
        data ["Close"] = float(close.replace(",",""))
    else:
        data ["Close"] = 0
    
    if dividend[1].get_text() != 'N/A' :
        data ["Open"] = float(dividend[1].get_text().replace(",",""))
    else:
        data ["Open"] = 0
    
    if dividend[6].get_text() != 'N/A' :
        data ["Volume"] = float(dividend[6].get_text().replace(",",""))
    else:
        data ["Volume"] = 0
        
    if dividend[7].get_text() != 'N/A' :    
        data ["Avg_Volume"] = float(dividend[7].get_text().replace(",",""))
    else:
        data ["Avg_Volume"] = 0
        
    data ["Market_cap"] = dividend[8].get_text()
    
    if dividend[10].get_text() != 'N/A' :
        data ["PE_Ratio"] = float(dividend[10].get_text().replace(",",""))
    else:
        data ["PE_Ratio"] = 0
    
    if dividend[11].get_text() != 'N/A' :
        data ["EPS"] = float(dividend[11].get_text().replace(",","")) 
    else:
        data ["EPS"] = 0
        
    data ["Dividend"] = dividend[13].get_text()
    
    if dividend[15].get_text() != 'N/A' :
        data ["Target_prize"] = float(dividend[15].get_text().replace(",","")) 
    else:
        data ["Target_prize"] = 0
        
    print(data)
        
    jsonFile.write(json.dumps(data, indent=3))
    jsonFile.write(",")
    jsonFile.flush()
jsonFile.write("]")
jsonFile.close()
f.close()
```
