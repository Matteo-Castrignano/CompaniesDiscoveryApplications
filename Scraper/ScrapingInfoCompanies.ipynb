```python
import re 
import json
import csv
from io import StringIO
from bs4 import BeautifulSoup
import requests
import yfinance as yf
```

```python
f = open("ListaSimboli1.txt", "r")
jsonFile = open("Companies.json", "a")
jsonFile.write("[")
i=1
for s in f: 
    
    stock= s.strip()

    try:
        companies = yf.Ticker(stock)
        dataCompanies = companies.info
    except:
        continue
        
    data = {}    
    data ["Symbol"] = stock
    
    try:
        data ["Name"] = dataCompanies["longName"]
    except:
        data ["Name"] = ""
        
    try:
        data ["Exchange"] = dataCompanies["exchange"]
    except:
        data ["Exchange"] = ""
        
    try:
        data ["Sector"] = dataCompanies["sector"]
    except:
        data ["Sector"] =""
        
    try:
        data ["FullTimesEmployees"] = dataCompanies["fullTimeEmployees"]
    except:
        data ["FullTimesEmployees"] = ""
        
    try:
        data ["Description"] = dataCompanies["longBusinessSummary"]
    except:
        data ["Description"] =""
        
    try:
        data ["City"] = dataCompanies["city"]
    except:
        data ["City"] = ""
        
    try:
        data ["Phone"] = dataCompanies["phone"]
    except:
        data ["Phone"] = ""
        
    try:
        data ["State"] = dataCompanies["state"]
    except:
        data ["State"] = ""
    
    try:
        data ["Country"] = dataCompanies["country"]
    except:
        data ["Country"] = ""
        
    try:
        data ["Address"] = dataCompanies["address1"]
    except:
        data ["Address"] = ""
        
    try:
        data ["Website"] = dataCompanies["website"]
    except:
        data ["Website"] = ""
        
    i=i+1
    print(i,data ["Symbol"], data ["Name"], data ["Exchange"])
        
    jsonFile.write(json.dumps(data, indent=3))
    jsonFile.write(",")
    jsonFile.flush()
jsonFile.write("]")
jsonFile.close()
f.close()
```
